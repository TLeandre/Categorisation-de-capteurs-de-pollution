# II- Préprocessing

### a) Données environnement capteurs 

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

##### 1_ Réglage des données mal convertit

```python
# nouveaux noms des colonnes 
env_colonnes = ['Capteurs',
            'Typologie_zone',
            'Type_voie','Nombre_voies',
            'Distance_capteur_voie',
            'Position_capteurs',
            'Presence_arbres',
            'Feuille_arbres',
            'Morphologie_urbaine',
            'Hauteur_batiments',
            'Distance_capteur_batiment',
            'Particularite',
            'Trafic_Routier_vehicule_leger',
            'Trafic_Routier_poids_lourds',
            'Trafic_Routier_2_roues',
            'emission_moyenne_annuelle_kg_maille_an',
            'Donnees_Cartographie']

# changement des noms de colonne
donnees_env.columns = env_colonnes

#Transformation de la colonne trafic routier vehicule léger en flotant et suppression des caractères inconnus 
donnees_env['Trafic_Routier_vehicule_leger'] = donnees_env['Trafic_Routier_vehicule_leger'].astype(str).replace('\xa0', '').apply(lambda x: float(str(x).replace('\xa0', '')))

# transformation des données des colonnes 12 à 16 en flottant ( afin d'avoir une unicité de type des variables ) 
colonnes_a_convertir = donnees_env.columns[13:16]
donnees_env[colonnes_a_convertir] = donnees_env[colonnes_a_convertir].astype(float)
```

##### 2_ Numérisation des variables qualitatives

```python
#fonction permettant de transformer les variables qualitatives en variables numériques 
def encodage_ordinal_colonne(df, min, max):
    # Récupération des colonnes à numériser 
    colonnes_a_encoder = df.columns[min:max]
    
    # Création d'un OrdinalEncoder
    ordinal_encoder = OrdinalEncoder()

    # Ajuster l'encodeur aux données et récupérer les catégories ordonnées
    df[colonnes_a_encoder] = ordinal_encoder.fit_transform(df[colonnes_a_encoder])
    
    return df

donnees_env = encodage_ordinal_colonne(donnees_env, 1, 12)
donnees_env
```

```python
donnees_env.info()
```

```python
## Suppression de la première colonne ( nom des capteurs ) 
capteurs = donnees_env['Capteurs']
donnees_env = donnees_env.drop(donnees_env.columns[0], axis=1)
```

##### 3_ Normalisation des données

```python
X = donnees_env.to_numpy()
Y = StandardScaler().fit_transform(X)

donnees_env_normees = pd.DataFrame(Y)
donnees_env_normees.insert(0, "Capteurs", capteurs)
donnees_env_normees.columns = env_colonnes

donnees_env_normees
```

### b) Données mesures

```python
date = donnees_mes['Date']
donnees_mes = donnees_mes.drop(donnees_mes.columns[0], axis=1)
```

##### 1_ Suppression des valeurs aberrantes

```python
for colonne in donnees_mes.columns:
        if pd.api.types.is_numeric_dtype(donnees_mes[colonne]):
            Q1 = donnees_mes[colonne].quantile(0.25)
            Q3 = donnees_mes[colonne].quantile(0.75)
            IQR = Q3 - Q1
            
            # Détection des valeurs aberrantes
            outliers_mask = (donnees_mes[colonne] < (Q1 - 1.5 * IQR)) | (donnees_mes[colonne] > (Q3 + 1.5 * IQR))
            
            # Remplacement des valeurs aberrantes par NaN
            donnees_mes.loc[outliers_mask, colonne] = np.nan
```

```python
donnees_mes.hist()
```

##### 2_ Suppression des colonnes avec un pourcentage de valeurs manquantes supérieur à 33%

```python
colonnes_del = []
for nom_colonne, nombre_na in nombre_na_par_colonne.items():
    if nombre_na / len(donnees_mes) > 0.33:
        colonnes_del.append([nom_colonne,nombre_na*100/len(donnees_mes)])
        donnees_mes = donnees_mes.drop(nom_colonne, axis=1)
colonnes_del
```

##### 3_ Remplissage des valeurs manquantes

```python
# Imputation par échantillonnage
for colonne in donnees_mes.columns:
    valeurs_manquantes = donnees_mes[colonne].isna()
    valeurs_disponibles = donnees_mes[colonne].dropna()
    imputation = np.random.choice(valeurs_disponibles, size=valeurs_manquantes.sum())
    donnees_mes.loc[valeurs_manquantes, colonne] = imputation

# Afficher les colonnes imputées
donnees_mes.hist()
```

##### 1_ Création des cercles de corrélations

```python
valeurs_propres = pca.explained_variance_ #valeurs propres                                       
vecteurs_propres = pca.components_ #vecteurs propres 

correlations = np.zeros((pca.n_components_, len(env_colonnes)-1))
for i in range(pca.n_components_):
    for y in range(len(env_colonnes)-1): # car on a garder index
        correlations[i,y] = np.sqrt(valeurs_propres[i]) * vecteurs_propres[i, y]
```

# VI- Clustering des données

### a) ACP

##### 1_ Classification ascendante hiérarchique

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import SpectralClustering
from sklearn.cluster import DBSCAN
import scipy.cluster.hierarchy as SCH

from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt
```

```python
algorithms = ["single", "complete", "average", "centroid", "ward"]
threshold = 8  # Le seuil de coupe du dendrogramme

# Boucle sur les algos
for algo in algorithms:
    # Créer une nouvelle figure pour chaque algorithme
    plt.figure(figsize=(5, 5))

    # Effectuer la CAH
    matrice_liaisons = SCH.linkage(Z_acp, method=algo, optimal_ordering=True)

    # Tracer le dendrogramme
    SCH.dendrogram(matrice_liaisons, color_threshold=threshold)

    # Ajouter un titre au dendrogramme
    plt.title(f"Dendrogramme - Algorithme: {algo}")

    # Afficher la figure
    plt.show()
```

```python
matrice_liaisons = SCH.linkage(Z_acp, method="ward", optimal_ordering=True)

# Essayer différents nombres de clusters
nombre_clusters = np.arange(5, 15, 0.5)
scores_silhouette = []

for n in nombre_clusters:
    cluster = SCH.fcluster(matrice_liaisons, n, criterion='distance')
    silhouette_avg = silhouette_score(Z_acp, cluster)
    scores_silhouette.append(silhouette_avg)

# Tracer la courbe de silhouette en fonction du nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(nombre_clusters, scores_silhouette, marker='o')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Score de Silhouette')
plt.title('Score de Silhouette en fonction du Nombre de Clusters')
plt.show()
```

```python
t=8
# Effectuez le clustering Classification ascendante hiérarchique avec une coupe à t = 8
matrice_liaisons = SCH.linkage(Z_acp, method="ward", optimal_ordering=True)
labels_CAH = SCH.fcluster(matrice_liaisons, t, criterion='distance')

plt.figure(figsize=(8, 6))
plt.scatter(Z_acp[:,0], Z_acp[:,1], c=labels_CAH, cmap='rainbow')
plt.title('Classification ascendante hiérarchique avec K=8')
plt.xlabel('Axe 1')
plt.ylabel('Axe 2')
plt.show()
```

##### 3_ GaussianMixture

```python
# Essayer différents nombres de clusters
nombre_clusters = range(1, 20)
bics = []
aics = []

for n in nombre_clusters:
    gmm = GaussianMixture(n_components=n)
    gmm.fit(Z_acp)
    # utilisation du Critère d'information d'Akaike et du Critère d'information bayésien
    bics.append(gmm.bic(Z_acp))
    aics.append(gmm.aic(Z_acp))

# Tracer les courbes BIC et AIC en fonction du nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(nombre_clusters, bics, marker='o', label='BIC')
plt.plot(nombre_clusters, aics, marker='o', label='AIC')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Critère d\'Information')
plt.title('Critères d\'Information en fonction du Nombre de Clusters')
plt.legend()
plt.show()
```

```python
# Essayer différents nombres de clusters
nombre_clusters = range(2, 20)
scores_silhouette = []

for n in nombre_clusters:
    gmm = GaussianMixture(n_components=n)
    labels = gmm.fit_predict(Z_acp)
    silhouette_avg = silhouette_score(Z_acp, labels)
    scores_silhouette.append(silhouette_avg)

# Tracer la courbe de silhouette en fonction du nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(nombre_clusters, scores_silhouette, marker='o')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Score de Silhouette')
plt.title('Score de Silhouette en fonction du Nombre de Clusters')
plt.show()
```

```python
# Effectuez le clustering GaussianMixture avec K=10
gmm = GaussianMixture(n_components=10)
# Obtenez les étiquettes des clusters
labels_gmm = gmm.fit_predict(Z_acp)
centres_clusters_gmm = gmm.means_

# Affichez les clusters en 2D
plt.figure(figsize=(8, 6))
plt.scatter(Z_acp[:, 0], Z_acp[:, 1], c=labels_gmm, cmap='rainbow')
plt.scatter(centres_clusters_gmm[:, 0], centres_clusters_gmm[:, 1], c='black', marker='x', s=100, label='Centres de clusters')
plt.title('GaussianMixture avec K=10')
plt.xlabel('Axe 1')
plt.ylabel('Axe 2')
plt.legend()
plt.show()
```

##### 4_ SpectralClustering

```python
# Essayer différents nombres de clusters
nombre_clusters = range(2, 20)
affinity = ['rbf','nearest_neighbors']

scores_silhouette_rbf = []
scores_silhouette_nn = []

for aff in affinity:
    for n_clusters in nombre_clusters:
        # Créer le modèle Spectral Clustering avec le nombre de clusters actuel
        modele_spectral = SpectralClustering(n_clusters=n_clusters, affinity=aff , random_state=42) 
        
         # Ajuster le modèle aux données
        labels = modele_spectral.fit_predict(Z_acp)
    
       # Calculer le coefficient de silhouette
        silhouette_coefficient = silhouette_score(Z_acp, labels)
        if aff == 'rbf':
            scores_silhouette_rbf.append(silhouette_coefficient)
        else :
            scores_silhouette_nn.append(silhouette_coefficient)
    
        # Afficher le résultat
        print(f"Nombre de clusters : {n_clusters}, Coefficient de silhouette : {silhouette_coefficient}")

# Tracer la courbe de silhouette en fonction du nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(nombre_clusters, scores_silhouette_rbf, marker='o', label='rbf')
plt.plot(nombre_clusters, scores_silhouette_nn, marker='o', label='nn')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Score de Silhouette')
plt.title('Score de Silhouette en fonction du Nombre de Clusters')
plt.legend()
plt.show()
```

```python
# Créer un modèle de Spectral Clustering
# n_clusters : nombre de clusters à former
spectral_clustering = SpectralClustering(n_clusters=9, affinity='nearest_neighbors', random_state=42)

# Obtenez les étiquettes des clusters
labels_SC = spectral_clustering.fit_predict(Z_acp)

# Calculer les centres des clusters (approximation)
centres_clusters_sc = np.array([Z_acp[labels_SC == i].mean(axis=0) for i in range(len(set(labels_SC)))])

# Affichez les clusters en 2D
plt.figure(figsize=(8, 6))
plt.scatter(Z_acp[:, 0], Z_acp[:, 1], c=labels_SC, cmap='rainbow')
plt.scatter(centres_clusters_sc[:, 0], centres_clusters_sc[:, 1], c='black', marker='x', s=100, label='Centres de clusters')
plt.title('Spectral Clustering avec K=10')
plt.legend()
plt.xlabel('Axe 1')
plt.ylabel('Axe 2')
plt.show()
```

##### 3_ GaussianMixture

```python
# Essayer différents nombres de clusters
nombre_clusters = range(1, 20)
bics = []
aics = []

for n in nombre_clusters:
    gmm = GaussianMixture(n_components=n)
    gmm.fit(Z_tsne)
    # utilisation du Critère d'information d'Akaike et du Critère d'information bayésien
    bics.append(gmm.bic(Z_tsne))
    aics.append(gmm.aic(Z_tsne))

# Tracer les courbes BIC et AIC en fonction du nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(nombre_clusters, bics, marker='o', label='BIC')
plt.plot(nombre_clusters, aics, marker='o', label='AIC')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Critère d\'Information')
plt.title('Critères d\'Information en fonction du Nombre de Clusters')
plt.legend()
plt.show()
```

```python
# Essayer différents nombres de clusters
nombre_clusters = range(2, 20)
scores_silhouette = []

for n in nombre_clusters:
    gmm = GaussianMixture(n_components=n)
    labels = gmm.fit_predict(Z_tsne)
    silhouette_avg = silhouette_score(Z_tsne, labels)
    scores_silhouette.append(silhouette_avg)

# Tracer la courbe de silhouette en fonction du nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(nombre_clusters, scores_silhouette, marker='o')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Score de Silhouette')
plt.title('Score de Silhouette en fonction du Nombre de Clusters')
plt.show()
```

##### 4_ Spectral Clustering

```python
# Essayer différents nombres de clusters
nombre_clusters = range(2, 20)
affinity = ['rbf','nearest_neighbors']

scores_silhouette_rbf = []
scores_silhouette_nn = []

for aff in affinity:
    for n_clusters in nombre_clusters:
        # Créer le modèle Spectral Clustering avec le nombre de clusters actuel
        modele_spectral = SpectralClustering(n_clusters=n_clusters, affinity=aff , random_state=42) 
        
         # Ajuster le modèle aux données
        labels = modele_spectral.fit_predict(Z_tsne)
    
       # Calculer le coefficient de silhouette
        silhouette_coefficient = silhouette_score(Z_tsne, labels)
        if aff == 'rbf':
            scores_silhouette_rbf.append(silhouette_coefficient)
        else :
            scores_silhouette_nn.append(silhouette_coefficient)
    
        # Afficher le résultat
        print(f"Nombre de clusters : {n_clusters}, Coefficient de silhouette : {silhouette_coefficient}")

# Tracer la courbe de silhouette en fonction du nombre de clusters
plt.figure(figsize=(10, 6))
plt.plot(nombre_clusters, scores_silhouette_rbf, marker='o', label='rbf')
plt.plot(nombre_clusters, scores_silhouette_nn, marker='o', label='nn')
plt.xlabel('Nombre de Clusters')
plt.ylabel('Score de Silhouette')
plt.title('Score de Silhouette en fonction du Nombre de Clusters')
plt.legend()
plt.show()
```

```python
# Créer un modèle de Spectral Clustering
# n_clusters : nombre de clusters à former
spectral_clustering = SpectralClustering(n_clusters=12, affinity='nearest_neighbors', random_state=42)

# Obtenez les étiquettes des clusters
labels_SC = spectral_clustering.fit_predict(Z_tsne)

# Calculer les centres des clusters (approximation)
centres_clusters_sc = np.array([Z_tsne[labels_SC == i].mean(axis=0) for i in range(len(set(labels_SC)))])

# Affichez les clusters en 2D
plt.figure(figsize=(8, 6))
plt.scatter(Z_tsne[:, 0], Z_tsne[:, 1], c=labels_SC, cmap='rainbow')
plt.scatter(centres_clusters_sc[:, 0], centres_clusters_sc[:, 1], c='black', marker='x', s=100, label='Centres de clusters')
plt.title('Spectral Clustering avec K=10')
plt.legend()
plt.xlabel('Axe 1')
plt.ylabel('Axe 2')
plt.show()
```

```python
# Faire varier les paramètres eps et min_samples
eps_values = [0.5, 1.0, 1.5, 2.0]
min_samples_values = [5, 10, 15, 20]

# Calculer le coefficient de silhouette pour chaque combinaison de paramètres
for eps in eps_values:
    for min_samples in min_samples_values:
        # Créer le modèle DBSCAN avec les paramètres actuels
        modele_dbscan = DBSCAN(eps=eps, min_samples=min_samples)

        # Ajuster le modèle aux données
        labels = modele_dbscan.fit_predict(Z_tsne)

        # Ignorer les clusters formés uniquement de points considérés comme du bruit (-1)
        if len(set(labels)) > 1:
            # Calculer le coefficient de silhouette
            silhouette_coefficient = silhouette_score(Z_tsne, labels)

            # Afficher le résultat
            print(f"eps={eps}, min_samples={min_samples}, Coefficient de silhouette : {silhouette_coefficient}")

```
